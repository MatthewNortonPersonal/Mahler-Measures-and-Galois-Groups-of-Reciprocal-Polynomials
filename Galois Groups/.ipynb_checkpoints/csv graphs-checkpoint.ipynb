### Function definitions

```python
def poly_bar_chart(df, data_color, limit, data_size, degree):
    df["Galois of trace"].value_counts().plot.bar(ylim=(0, limit), color = data_color, ylabel = "number of polynomials")
    plt.text(0.97, 0.96, "d = " + str(degree), fontsize = 16, ha = "right", va = "top", transform=plt.gca().transAxes)
    plt.text(0.98, 0.84, str(data_size) + " polynomials analyzed", ha = "right", va = "center", transform=plt.gca().transAxes)
```
